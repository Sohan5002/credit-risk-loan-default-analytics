```python
#Gender Analysis
```

```python
gender_analysis = (
    df.groupby("gender")
      .agg(
          Total_Customers=("applicant_id", "count"),
          Total_Loan_Amount=("loan_amount", "sum"),
          Average_Loan=("loan_amount", "mean"),
          Default_Rate=("default_flag", "mean")
      )
)

gender_analysis["Default_Rate"] = (
    gender_analysis["Default_Rate"] * 100
).round(2)

gender_analysis
```

```python
plt.figure(figsize=(6,4))

sns.countplot(
    data=df,
    x="gender"
)

plt.title("Gender Distribution")
plt.show()
```

```python
#Age Group Analysis

```

```python
age_analysis = (
    df.groupby("age_group")
      .agg(
          Total_Customers=("applicant_id","count"),
          Average_Loan=("loan_amount","mean"),
          Default_Rate=("default_flag","mean")
      )
)

age_analysis["Default_Rate"] = (
    age_analysis["Default_Rate"] * 100
).round(2)

age_analysis

```

```python
#Marital Status Analysis
marital_analysis = (
    df.groupby("marital_status")
      .agg(
          Total_Customers=("applicant_id","count"),
          Average_Loan=("loan_amount","mean"),
          Default_Rate=("default_flag","mean")
      )
)

marital_analysis["Default_Rate"] = (
    marital_analysis["Default_Rate"] * 100
).round(2)

marital_analysis
```

```python
# Education Analysis
education_analysis = (
    df.groupby("education")
      .agg(
          Total_Customers=("applicant_id","count"),
          Average_Loan=("loan_amount","mean"),
          Default_Rate=("default_flag","mean")
      )
)

education_analysis["Default_Rate"] = (
    education_analysis["Default_Rate"] * 100
).round(2)

education_analysis
```

```python
plt.figure(figsize=(10,5))

sns.countplot(
    data=df,
    x="education",
    order=df["education"].value_counts().index
)

plt.xticks(rotation=45)
plt.title("Education Distribution")
plt.show()
```

```python
#Employment Type Analysis
```

```python
employment_analysis = (
    df.groupby("employment_type")
      .agg(
          Total_Customers=("applicant_id","count"),
          Average_Loan=("loan_amount","mean"),
          Default_Rate=("default_flag","mean")
      )
)

employment_analysis["Default_Rate"] = (
    employment_analysis["Default_Rate"] * 100
).round(2)

employment_analysis
```

```python
plt.figure(figsize=(8,5))

sns.countplot(
    data=df,
    x="employment_type",
    order=df["employment_type"].value_counts().index
)

plt.xticks(rotation=30)
plt.title("Employment Type Distribution")
plt.show()
```

```python
#Credit Score vs Default
credit_analysis = (
    df.groupby("credit_score_category")
      .agg(
          Customers=("loan_id", "count"),
          Default_Rate=("default_flag", "mean")
      )
      .reset_index()
)

credit_analysis["Default_Rate"] = (
    credit_analysis["Default_Rate"] * 100
).round(2)

credit_analysis
```

```python
plt.figure(figsize=(8,5))

sns.barplot(
    data=credit_analysis,
    x="credit_score_category",
    y="Default_Rate"
)

plt.title("Credit Score Category vs Default Rate")
plt.ylabel("Default Rate (%)")

plt.show()
```

```python
#Debt-to-Income Ratio Analysis
dti_analysis = (
    df.groupby("high_risk_flag")
      .agg(
          Customers=("loan_id", "count"),
          Average_DTI=("debt_to_income_ratio", "mean"),
          Default_Rate=("default_flag", "mean")
      )
      .reset_index()
)

dti_analysis["Default_Rate"] = (
    dti_analysis["Default_Rate"] * 100
).round(2)

dti_analysis
```

```python
plt.figure(figsize=(6,4))

sns.barplot(
    data=dti_analysis,
    x="high_risk_flag",
    y="Default_Rate"
)

plt.xticks([0, 1], ["Low Risk", "High Risk"])

plt.title("High Risk Flag vs Default Rate")
plt.ylabel("Default Rate (%)")

plt.show()
```

```python
#Default Rate by Income Category
plt.figure(figsize=(6,4))

sns.barplot(
    data=dti_analysis,
    x="high_risk_flag",
    y="Default_Rate"
)

plt.xticks([0, 1], ["Low Risk", "High Risk"])

plt.title("High Risk Flag vs Default Rate")
plt.ylabel("Default Rate (%)")

plt.show()
```

```python
existing_loan_analysis = (
    df.groupby("loan_count_category")
      .agg(
          Customers=("loan_id", "count"),
          Default_Rate=("default_flag", "mean")
      )
      .reset_index()
)

existing_loan_analysis["Default_Rate"] = (
    existing_loan_analysis["Default_Rate"] * 100
).round(2)

existing_loan_analysis

plt.figure(figsize=(8,5))

sns.barplot(
    data=existing_loan_analysis,
    x="loan_count_category",
    y="Default_Rate"
)

plt.title("Existing Loans vs Default Rate")
plt.ylabel("Default Rate (%)")
plt.show()
```

```python
#Employment Type vs Default
employment_default = (
    df.groupby("employment_type")
      .agg(
          Customers=("loan_id", "count"),
          Default_Rate=("default_flag", "mean"),
          Average_Credit_Score=("credit_score", "mean")
      )
      .reset_index()
)

employment_default["Default_Rate"] = (
    employment_default["Default_Rate"] * 100
).round(2)

employment_default
```

```python
plt.figure(figsize=(8,5))

sns.barplot(
    data=employment_default,
    x="employment_type",
    y="Default_Rate"
)

plt.title("Employment Type vs Default Rate")
plt.ylabel("Default Rate (%)")
plt.show()
```

```python
#Loan Product vs Default
product_default = (
    df.groupby("loan_product")
      .agg(
          Customers=("loan_id", "count"),
          Default_Rate=("default_flag", "mean")
      )
      .reset_index()
)

product_default["Default_Rate"] = (
    product_default["Default_Rate"] * 100
).round(2)

product_default = product_default.sort_values(
    by="Default_Rate",
    ascending=False
)

product_default
```

```python
plt.figure(figsize=(10,5))

sns.barplot(
    data=product_default,
    x="loan_product",
    y="Default_Rate"
)

plt.xticks(rotation=45)
plt.title("Loan Product vs Default Rate")
plt.ylabel("Default Rate (%)")
plt.show()
```
