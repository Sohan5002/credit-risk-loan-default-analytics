```python
top_age = df["age_group"].mode()[0]
count = df["age_group"].value_counts().iloc[0]
percentage = df["age_group"].value_counts(normalize=True).iloc[0] * 100

print("📌 Business Insight")

print(f"""
The '{top_age}' age group has the highest number of customers ({count:,}).

It contributes {percentage:.2f}% of the total customer portfolio.
""")
```
