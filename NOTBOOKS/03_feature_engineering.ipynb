```python
import pandas as pd
import numpy as np

df = pd.read_csv("../DATA/cleaned_loan_data.csv")
```

```python
feature_df = df.copy()
```

```python
feature_df.shape
```

```python
feature_df["age_group"] = pd.cut(
    feature_df["age"],
    bins=[18,25,35,45,55,65,100],
    labels=["18-25","26-35","36-45","46-55","56-65","65+"]
)
```

```python
import os

os.makedirs("../DATA/processed", exist_ok=True)

feature_df.to_csv(
    "../DATA/processed/loan_feature_engineered.csv",
    index=False
)

print("File Saved Successfully!")
```

```python
feature_df["income_category"] = pd.qcut(
    feature_df["annual_income"],
    q=4,
    labels=["Low", "Medium", "High", "Very High"]
)

feature_df["income_category"].value_counts()
```

```python
feature_df["credit_score_category"] = pd.cut(
    feature_df["credit_score"],
    bins=[300, 580, 670, 740, 800, 900],
    labels=["Poor", "Fair", "Good", "Very Good", "Excellent"]
)

feature_df["credit_score_category"].value_counts()
```

```python
feature_df["loan_amount_category"] = pd.qcut(
    feature_df["loan_amount"],
    q=4,
    labels=["Small", "Medium", "Large", "Very Large"]
)

feature_df["loan_amount_category"].value_counts()
```

```python
feature_df["emi_to_income_ratio"] = (
    feature_df["emi_amount"] / feature_df["monthly_income"]
).round(3)

feature_df["emi_to_income_ratio"].describe()
```

```python
feature_df["loan_utilization"] = (
    feature_df["disbursed_amount"] /
    feature_df["sanctioned_amount"]
).round(3)

feature_df["loan_utilization"].describe()
```

```python
feature_df["approval_percentage"] = (
    feature_df["sanctioned_amount"] /
    feature_df["loan_amount"] * 100
).round(2)

feature_df["approval_percentage"].describe()
```

```python
feature_df["high_risk_flag"] = np.where(
    (feature_df["credit_score"] < 650) &
    (feature_df["debt_to_income_ratio"] > 40),
    1,
    0
)

feature_df["high_risk_flag"].value_counts()
```

```python
feature_df["loan_count_category"] = pd.cut(
    feature_df["existing_loans_count"],
    bins=[-1, 0, 2, 5, 20],
    labels=["No Loan", "1-2", "3-5", "5+"]
)

feature_df["loan_count_category"].value_counts()
```

```python
feature_df["default_flag"] = np.where(
    feature_df["loan_status"].isin(["Default", "Written-Off"]),
    1,
    0
)

feature_df["default_flag"].value_counts()
```
