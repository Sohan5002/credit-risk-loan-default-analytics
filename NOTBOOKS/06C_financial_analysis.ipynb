```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("ggplot")
sns.set_theme(style="whitegrid")

pd.set_option("display.max_columns", None)
```

```python
df = pd.read_csv("FinEdge_Capital_Loan_Portfolio.csv")
```

```python
df.head()
```

```python
df.columns.tolist()
```

```python
financial_summary = pd.DataFrame({

    "Metric":[
        "Total Loans",
        "Total Loan Amount",
        "Total Sanctioned Amount",
        "Total Disbursed Amount",
        "Total Recovery Amount",
        "Average Interest Rate (%)",
        "Average EMI",
        "Average Processing Fee",
        "Average Approval Percentage (%)"
    ],

    "Value":[
        len(df),
        df["loan_amount"].sum(),
        df["sanctioned_amount"].sum(),
        df["disbursed_amount"].sum(),
        df["recovery_amount"].sum(),
        df["interest_rate"].mean(),
        df["emi_amount"].mean(),
        df["processing_fee"].mean(),
        df["approval_percentage"].mean()
    ]

})

financial_summary
```

```python
financial_summary["Value"] = financial_summary["Value"].apply(lambda x: f"{x:,.2f}")

financial_summary
```

```python
print(f"Total Loan Amount : ₹ {df['loan_amount'].sum():,.2f}")
```

```python
print(f"Total Disbursed Amount : ₹ {df['disbursed_amount'].sum():,.2f}")
```

```python
print(f"Total Recovery Amount : ₹ {df['recovery_amount'].sum():,.2f}")
```

```python
comparison = pd.DataFrame({

    "Type":["Loan Amount","Disbursed Amount"],

    "Amount":[
        df["loan_amount"].sum(),
        df["disbursed_amount"].sum()
    ]

})

comparison
```

```python
plt.figure(figsize=(6,5))

sns.barplot(
    data=comparison,
    x="Type",
    y="Amount"
)

plt.title("Loan Amount vs Disbursed Amount")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.histplot(df["approval_percentage"], bins=20)

plt.title("Approval Percentage Distribution")

plt.show()
```

```python
recovery = df.groupby("loan_status")["recovery_amount"].sum().sort_values(ascending=False)

recovery
```

```python
plt.figure(figsize=(8,5))

sns.barplot(
    x=recovery.index,
    y=recovery.values
)

plt.title("Recovery Amount by Loan Status")

plt.xticks(rotation=30)

plt.show()
```
